### 2.1 

Let $C ‚äÜ R^n$ be a convex set, with $x_{1},...,x_{k} \in C$, and let $Œ∏_{1},...,Œ∏_{k}\in R$ satisfy $Œ∏_{i} \geq 0$, $Œ∏_{1} +¬∑¬∑¬∑+Œ∏_{k} = 1$. Show that $Œ∏_{1}x_{1} +¬∑¬∑¬∑+Œ∏_{k}x_{k} \in C$. (The definition of convexity is that this holds for $k = 2$; you must show it for arbitrary k.) **Hint. Use induction on k.**
#### Solution:

The proof is by induction on k, the number of terms in the convex combination.

When $k = 1$ , this just says that each point of $C$ is a point of $C$. When $k = 2$, the statement of the
theorem is the definition of a convex set, the set of convex combinations a set $C$ is convex if the line segment between any two points in $C$ lies in $C$, i.e., if for any $x_{1}, x_{2} \in C$ and any $Œ∏$ with $0 \leq Œ∏ \leq 1$, we have $Œ∏x_{1} +(1‚àíŒ∏)x_{2} \in C$. 

Now assume all length (k ‚àí 1) combinations are contained in $C$, and take a length k combination of points in $C$: 

\begin{equation} 
Œ∏_{1}x_{1} + Œ∏_{2}x_{2}+¬∑¬∑¬∑+Œ∏_{k}x_{k}
\end{equation}

By the inductive hypothesis, we know that

\begin{equation} 
y =\frac{ Œ∏_{1}}{Œ∏_{1} +...+ Œ∏_{k-1}} x_{1} + \frac{ Œ∏_{2}}{Œ∏_{1} +...+ Œ∏_{k-1}}x_{2}+¬∑¬∑¬∑+\frac{ Œ∏_{k-1}}{Œ∏_{1} +...+ Œ∏_{k-1}}x_{k-1} \in C
\end{equation}

(This is only defined if $Œ∏_{1}+Œ∏_{2}...+Œ∏_{k-1} \neq 0$; but if it‚Äôs $0$, then $Œ∏_{k}$ is the only nonzero
coefficient, so we effectively had a length-1 convex combination to begin with.) But now, the
original convex combination can be written as: 

\begin{equation} 
Œ∏_{1}x_{1} + Œ∏_{2}x_{2}+¬∑¬∑¬∑+Œ∏_{k}x_{k} = (Œ∏_{1}+ Œ∏_{2}+¬∑¬∑¬∑+Œ∏_{k})y +Œ∏_{k}x_{k}
\end{equation}

which lies on the line segment $[y,x_{k}]$,and therefore it is in $C$ by the definition of a convex set.
By induction, convex combinations of all size must be contained in $C$.

### 2.3

Midpoint convexity. A set C is midpoint convex if whenever two points a,b are in C, the average or midpoint (a + b)/2 is in C. Obviously a convex set is midpoint convex. It can be proved that under mild conditions midpoint convexity implies convexity. As a simple case, prove that if C is closed and midpoint convex, then C is convex.

#### Solution

First we have to show that $Œ∏x_{1} +(1-Œ∏)x_{2}\in C, \forall Œ∏ \in [0,1]$ and $x_{1},x_{2} \in C$ 

We know that

\begin{equation} 
 \frac{ x_{1}+x_{2}}{2} x_{1} \in C \Longrightarrow \frac{ x_{1}+\frac{ x_{1}+x_{2}}{2}}{2} = \frac{3}{4} x_{1} + \frac{1}{4}x_{2} \in C
\end{equation}

Let $Œ∏_{k}$ be the binary number of length k, i.e., a number of the form

\begin{equation} 
 Œ∏_{k} =c_{1}2^{‚àí1} +c_{2}2^{‚àí2} +¬∑¬∑¬∑+c_{k}2^{‚àík}

### 2.4

Show that the convex hull of a set $S$ is the intersection of all convex sets that contain $S$. (The same method can be used to show that the conic, or affine, or linear hull of a set $S$ is the intersection of all conic sets, or affine sets, or subspaces that contain $S$.)

#### Solution

Let $C$ be the convex hull of $S$ and let $P$ be the intersection of all convex sets that contain $S$, i.e.,

\begin{equation} 
P = \bigcap \left\lbrace P|P convex, S\subseteq P\right \rbrace
\end{equation}

We will show that $C = P$ by showing that $C ‚äÜ P$ and $P ‚äÜ C$.

First we show that $C ‚äÜ P$.
Suppose $x ‚àà C$, i.e., x is a convex combination of some points $x_{1},...,x_{k} ‚àà S$. Now let $P$ be any convex set such that $S ‚äÜ P$. Evidently, we have $x_{1},...,x_{n} ‚àà P$. Since $P$ is convex, and x is a convex combination of $x_{1},...,x_{k}$, it follows that $x ‚àà P$. We have shown that for any convex set $P$ that contains $S$, we have $x ‚àà P$. This means that $x$ is in the intersection of all convex sets that contain $S$, i.e., $x ‚àà P$. Now let us show that P ‚äÜ C. Since C is convex (by definition) and contains $S$, we must have $C = P$ for some $P$ in the construction of $P$, proving the claim.

### 2.13

Conic hull of outer products. Consider the set of rank-k outer products, defined as {$XX^T | X ‚àà R^{n√ók}, rank X = k$}. Describe its conic hull in simple terms.
 
#### Solution. 

We have $XX^T ‚âΩ 0$ and $rank(XX^T ) = k$. A positive combination of such matrices can have rank up to $n$, but never less than $k$. Indeed, Let $A$ and $B$ be positive semidefinite matrices of rank $k$, with $rank(A + B) = r < k$. Let $V ‚àà R^{n√ó(n‚àír)}$ be a matrix with $R(V ) = N (A + B)$, i.e.,

$$V^T(A+B)V =V^TAV +V^TBV =0$$

Since $A, B ‚âΩ 0$, this means
$$V^TAV =V^TBV =0$$

which implies that $rank A ‚â§ r$ and $rank B ‚â§ r$. We conclude that $rank(A + B) ‚â• k$ for any $A$,$B$ such that $rank(A,B) = k$ and $A$,$B ‚âΩ 0$, It follows that the conic hull of the set of rank-k outer products is the set of positive semidefinite matrices of rank greater than or equal to $k$, along with the zero matrix.
